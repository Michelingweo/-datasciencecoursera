```python
My_table = soup.find('table',{'class':'wikitable sortable'})
```

```python
links=My_table.findAll('a')
```

```python
links2=My_table.findAll('td')
```

```python
Borough=[]
for link in links:
    Borough.append(link.get('title'))
Post=[]
for link in links2:
    Post.append(link.get('title'))
```

```python
codes_list=[]
borough_list=[]
neighborhood_list=[]
i=1
for tag in My_table.find_all('td'):
    if i == 1:
        codes_list.append(tag.text)
    if i == 2:
        borough_list.append(tag.text)
    if i == 3: 
        neighborhood_list.append(tag.text)
    i = i+1
    if i==4:
        i=1
```
